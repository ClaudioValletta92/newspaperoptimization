```python
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
from nltk.sentiment import SentimentIntensityAnalyzer
from gensim import corpora, models
from pprint import pprint
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import SnowballStemmer
from gensim import corpora, models
from pprint import pprint
def download_title(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract the article title based on the HTML structure
        # Modify this part based on the specific structure of the webpage
        title_element = soup.find('h1', class_='fp_intro__title')

        # Print or process the title as needed
        if title_element:
            title_text = title_element.get_text(strip=True)
            return title_text
        else:
            return "-"

    else:
        return "-"


def download_abstract(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract the article abstract based on the HTML structure
        abstract_element = soup.find('div', class_='fp_intro__abstract')
        if abstract_element:
            abstract_text = abstract_element.get_text(strip=True)
            return abstract_text
        else:
            print("Couldn't find the abstract on the webpage.")

    else:
        print(f"Failed to fetch the webpage. Status Code: {response.status_code}")
def download_shares(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract the number of shares based on the HTML structure
        shares_element = soup.find('div', class_='fp_share__counter')
        if shares_element:
            shares_text = shares_element.find('span').get_text(strip=True)
            return shares_text
        else:
            print("Couldn't find the number of shares on the webpage.")

    else:
        print(f"Failed to fetch the webpage. Status Code: {response.status_code}")
def download_citations(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract citations based on the HTML structure
        citation_elements = soup.find_all('blockquote')
        if citation_elements:
            for index, citation_element in enumerate(citation_elements, start=1):
                citation_text = citation_element.get_text(strip=True)
                print(f"Citation {index}: {citation_text}")
        else:
            print("Couldn't find any citations on the webpage.")

    else:
        print(f"Failed to fetch the webpage. Status Code: {response.status_code}")
def download_subtitles(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract subtitles based on the HTML structure
        subtitle_elements = soup.find_all('h2')
        if subtitle_elements:
            for index, subtitle_element in enumerate(subtitle_elements, start=1):
                subtitle_text = subtitle_element.get_text(strip=True)
                print(f"Subtitle {index}: {subtitle_text}")
        else:
            print("Couldn't find any subtitles on the webpage.")

    else:
        print(f"Failed to fetch the webpage. Status Code: {response.status_code}")
def download_paragraph_count(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract all text within <p> tags based on the HTML structure
        paragraph_elements = soup.find_all('p')
        if paragraph_elements:
            return len(paragraph_elements)
        #    for index, paragraph_element in enumerate(paragraph_elements, start=1):
                # Use .text instead of .get_text() to handle text within tags
        #        paragraph_text = ' '.join([part.text if part.name != 'a' else part.get_text(strip=True) for part in paragraph_element.contents])
         #       paragraph_text = paragraph_text.replace('<strong>', '').replace('</strong>', '')
          #      print(f"Paragraph {index}: {paragraph_text} ")
        else:
            print("Couldn't find any paragraphs on the webpage.")

    else:
        print(f"Failed to fetch the webpage. Status Code: {response.status_code}")

def download_article_text(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract all text within <p> tags based on the HTML structure
        paragraph_elements = soup.find_all('p')
        article_text = ''

        if paragraph_elements:
            for paragraph_element in paragraph_elements:
                # Use .text instead of .get_text() to handle text within tags
                paragraph_text = ' '.join([part.text if part.name != 'a' else part.get_text(strip=True) for part in paragraph_element.contents])

                # Concatenate paragraph text to the article text
                article_text += paragraph_text + '\n'

            return article_text.strip()  # Remove leading and trailing whitespaces
        else:
            print("Couldn't find any paragraphs on the webpage.")
            return None

    else:
        print(f"Failed to fetch the webpage. Status Code: {response.status_code}")
        return None
def download_date(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract the date based on the HTML structure
        date_element = soup.find('time', class_='fp_info__date-updated')
        if date_element:
            date_text = date_element['datetime']
            return date_text
        else:
            return "-"

    else:
        return "-"
def download_author(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract the author name based on the HTML structure
        author_element = soup.find('a', attrs={'data-anact': 'autore'})
        if author_element:
            author_name = author_element.get_text(strip=True)
            return author_name
        else:
            return "-"

    else:
        return "-"

def count_hyperlinks_in_paragraphs(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract all paragraphs
        paragraph_elements = soup.find_all('p')
        total_hyperlink_count=0
        # Count hyperlinks in each paragraph
        for index, paragraph_element in enumerate(paragraph_elements, start=1):
            hyperlink_count = len(paragraph_element.find_all('a'))
            total_hyperlink_count=total_hyperlink_count+hyperlink_count
    else:
        print(f"Failed to fetch the webpage. Status Code: {response.status_code}")
    return total_hyperlink_count

def count_strong_tags_in_paragraphs(url):
    # Send a GET request to the URL
    response = requests.get(url)
    strong_total_count=0
    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract all paragraphs
        paragraph_elements = soup.find_all('p')

        # Count <strong> tags within each paragraph
        for index, paragraph_element in enumerate(paragraph_elements, start=1):
            strong_count = len(paragraph_element.find_all('strong'))
            strong_total_count=strong_total_count+strong_count

    else:
        print(f"Failed to fetch the webpage. Status Code: {response.status_code}")
    return strong_total_count
def count_em_tags_in_paragraphs(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract all paragraphs
        paragraph_elements = soup.find_all('p')

        # Count <strong> tags within each paragraph
        for index, paragraph_element in enumerate(paragraph_elements, start=1):
            strong_count = len(paragraph_element.find_all('em'))
            print(f"Paragraph {index} <em> Tag Count: {strong_count}")

    else:
        print(f"Failed to fetch the webpage. Status Code: {response.status_code}")
def find_elements_around_li(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all <li> tags in the document
        li_elements = soup.find_all('li')

        # Find and print all elements around each <li> tag
        for index, li_element in enumerate(li_elements, start=1):
            # Find all parents of the <li> tag
            parents = li_element.find_parents()

            # Print the information for each parent
            for parent_index, parent in enumerate(parents, start=1):
                print(f"Element {parent_index} around <li> in <li> {index}:\n{parent}\n")

    else:
        print(f"Failed to fetch the webpage. Status Code: {response.status_code}")

def compute_average_words_per_paragraph(paragraph,words):
    return words/paragraph

def compute_text_length(text):
    # Split the text into words using whitespace as the delimiter
    words = text.split()

    # Return the number of words in the text
    return len(words)    

def remove_italian_stopwords(text):
    # Tokenize the text into words
    words = word_tokenize(text, language='italian')

    # Get the Italian stopwords
    italian_stopwords = set(stopwords.words('italian'))

    # Remove stopwords from the list of words
    filtered_words = [word for word in words if word.lower() not in italian_stopwords]

    # Join the filtered words back into a string
    filtered_text = ' '.join(filtered_words)

    return filtered_text

def analyze_sentiment_vader_italian(text):
    # Create a SentimentIntensityAnalyzer object
    sia = SentimentIntensityAnalyzer()

    # Get sentiment scores (positive, neutral, negative, and a compound score)
    sentiment_scores = sia.polarity_scores(text)

    # Extract the compound score
    compound_score = sentiment_scores['compound']
    positive_score= sentiment_scores['pos']
    negative_score= sentiment_scores['neg']
    neutral_score= sentiment_scores['neu']

    sentiment_dictionary={'pos':positive_score,'neg':negative_score,'neut':neutral_score,'compound':compound_score}

    return sentiment_dictionary

def preprocess_text(text, language='italian'):
    stop_words = set(stopwords.words(language))
    stemmer = SnowballStemmer(language)

    tokens = word_tokenize(text.lower())
    tokens = [stemmer.stem(token) for token in tokens if token.isalnum() and token not in stop_words]

    return tokens

def extract_topics_with_labels(texts, num_topics=5):
    tokenized_texts = [preprocess_text(text) for text in texts]

    dictionary = corpora.Dictionary(tokenized_texts)
    corpus = [dictionary.doc2bow(tokens) for tokens in tokenized_texts]

    lda_model = models.LdaModel(corpus, num_topics=num_topics, id2word=dictionary)
    topics = lda_model.print_topics(num_topics=num_topics, num_words=5)

    topic_labels = []
    for topic in topics:
        topic_id, terms = topic
        term_list = [term.split("*")[1].strip('"') for term in terms.split(" + ")]
        label = ", ".join(term_list)
        topic_labels.append((topic_id, label))

    return topic_labels

def extract_author_date_and_date(url):
    print(url)
    # Add your logic here to extract author name and date based on the URL
    # For demonstration purposes, I'll return placeholder values
    url_mod='https://www.fanpage.it'+url
    author_name=download_author(url_mod)
    title=download_title(url_mod)
    date=download_date(url_mod)
    return author_name,title,date
```

```python
daily_summary_df = pd.DataFrame()

# Group by 'Date' and calculate sum of views and count of articles
daily_summary_df = notiziedf_2023.groupby(notiziedf_2023['Date'].dt.date).agg(
    Views=pd.NamedAgg(column='Views', aggfunc='sum'),
    Total_Articles=pd.NamedAgg(column='Date', aggfunc='size')
).reset_index()

daily_summary_df['Date'] = pd.to_datetime(daily_summary_df['Date'])  # Convert date back to datetime
daily_summary_df['Views_per_Article'] = daily_summary_df['Views'] / daily_summary_df['Total_Articles']


# Display the resulting DataFrame
print(daily_summary_df)
```

```python
keywords = ['quando|quand|quand’', 'perche|perché','chi','cosa|cos’|cos','dove|dov|dov’']

# Create new columns indicating whether the title contains each specified keyword
for keyword in keywords:
    notiziedf_2023[f'Contains_{keyword.replace("|", "_")}'] = notiziedf_2023['title'].str.contains(keyword, case=False, regex=True)

# Group by the new columns
grouped_data_keywords = notiziedf_2023.groupby([f'Contains_{keyword.replace("|", "_")}' for keyword in keywords]).agg({'Pagina': 'count', 'Views': 'sum'}).reset_index()

# Divide 'Views' by one million
grouped_data_keywords['Views'] = grouped_data_keywords['Views'] / 1000000

# Rename columns for clarity
grouped_data_keywords.columns = [f'Contains_{keyword.replace("|", "_")}' for keyword in keywords] + ['TotalArticles', 'TotalViews']

# Calculate average views per article
grouped_data_keywords['AverageViews'] = grouped_data_keywords['TotalViews'] * 1000000 / grouped_data_keywords['TotalArticles']

# Print the result
print("Grouped Data by Keywords in Title:")
grouped_data_keywords
```
